## Inferring Cell Type
Inferring the cell type of the cancer cell line used in the myllia perturbation dataset

```python
import scanpy as sc
import infercnvpy as cnv
```

```python
adata = sc.read_h5ad("/home/data/kaggle_data/myllia/training_cells.h5ad")

# basic preprocessing
sc.pp.normalize_total(adata)
sc.pp.log1p(adata)

sc.pp.highly_variable_genes(adata)
adata = adata[:, adata.var.highly_variable]

sc.pp.scale(adata)
sc.tl.pca(adata)
sc.pp.neighbors(adata)
sc.tl.umap(adata)
```
